```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
#I am initializing my dataframes to python
df_tsla = pd.read_csv("TSLA.csv")
df_tweet = pd.read_csv("TweetsElonMusk.csv")
```

```python
#Here I am establishing a column for "Date", so that the two dataframes can be merged
df_tweet["Date"] = df_tweet["date"]
```

```python
#Here I am merging the datasets for the stock price of tesla and the date of the tweets
df_tweet_tsla_price = pd.merge(df_tsla, df_tweet, on = "Date")
df_tweet_tsla_price
```

```python
#Here I am grouping the days that Elon Musk tweets something, I am including the count method so that the dates may be grouped 
num_of_tweets_in_day = df_tweet_tsla_price.groupby("Date").count()
num_of_tweets_in_day
```

```python
#Here I am combining the stock price of telsa to days that Elon Musk tweets something
df_tsla_numtwts = pd.merge(num_of_tweets_in_day, df_tsla, on = "Date")
df_tsla_numtwts
```

```python
#Here I am changing the column names, Open_x originally was the number of tweets, I chaged the name to "Number of Tweets" for readability, I then created the column price change by subtracting the closing price from the opening price
df_tsla_numtwts["Number of tweets"] = df_tsla_numtwts["Open_x"]
df_tsla_numtwts["Price Change"] = df_tsla_numtwts["Close_y"] - df_tsla_numtwts["Open_y"]
df_tsla_numtwts
```

```python
#Here I am creating a dataset with all the variables that I want to use for the final inference, on line 3, I am renaming the columns to have more appealing names
FinalDF = df_tsla_numtwts[["Date","Open_y","High_y","Low_y","Close_y","Adj Close_y","Volume_y","Number of tweets","Price Change"]]
FinalDF.columns = ["Date","Open","High","Low","Close","Adj_Close","Volume","Number of Tweets", "Price Change"]
FinalDF["Pos_Neg"] = ""
FinalDF
```

```python
#Here I am establishing a counter to count the numbers of positive, negative, and no change in stock price based on days that Elon Musk tweeted 
pos_count = 0
neg_count = 0
zero_count = 0
#Here I establish a variable for the change of price 
price_change = FinalDF["Price Change"]
#I create an empty list to create the categorical variable
pos_neg = []
#I am creating a for loop which establishes the categories of price change which are Zero, positive, and negative
for i in price_change:
    if i > 0:
        pos_neg.append("Pos")
        pos_count += 1
    elif i < 0:
        pos_neg.append("Neg")
        neg_count += 1
    else:
        pos_neg.append("Zero")
        zero_count += 1
FinalDF["Pos_Neg"] = pos_neg

#Here I am finding the percentages of stock price changes 
percent_positive = pos_count/len(pos_neg)
percent_negative = neg_count/ len(pos_neg)
percent_zero = zero_count/len(pos_neg)
```

```python
#This will print the summary statistics for the price change of the Tesla stock price, on days where Elon Musk tweets
FinalDF["Price Change"].describe()
```

```python
#This will print the summary statistics for the number of tweets by Elon Musk, on days that he has at least one tweet
FinalDF["Number of Tweets"].describe()
```

```python
#Here is the average price change of Tesla, on days that he tweets 
sns.distplot(FinalDF['Price Change'])
```

```python
#This graph is for the average number of tweets by Elon Musk, on days that he tweets
sns.distplot(FinalDF['Number of Tweets'])
```

```python
#Here I am creating a joint plot, which compares the number of tweets to the change of tesla price
sns.jointplot(x=FinalDF['Number of Tweets'],
             y=FinalDF['Price Change'], kind = 'reg', color = "green")
#I chose to go with green because there is a positive correlation, the top graph is the average number of tweets and the left most graph is the average price change
```

```python
#Here I am creating a joint plot, which compares the number of tweets to the volume Tesla is traded at
sns.jointplot(x=FinalDF['Number of Tweets'],
             y=FinalDF['Volume'], kind = 'reg', color = "green")
#I chose to go with green because there is a positive correlation, the top graph is the average number of tweets and the left most graph is the average volume of Tesla trades
```

```python
#Here I initialize a dictionary to utilize the pie chart code that was used in class
percent_dict = {}
percent_dict["Positive"] = percent_positive
percent_dict["Negative"] = percent_negative
percent_dict["Zero"] = percent_zero

plt.pie([x*100 for x in percent_dict.values()],
       labels = [x for x in percent_dict.keys()],
       autopct = '%0.1f', explode = [0, 0, 1])

# label
plt.title("Change in Price")
plt.savefig('pieplot')

plt.show()
```
